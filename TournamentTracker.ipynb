```python
#import csv package to save the list as csv file
import csv
#Used to print out the string and a line of "=" underneath the string
def separator(string):
    print(string + '\n' + '=' * len(string))                   
```

```python
def sign_Up(aList):
    separator('Participant Sign Up')
    #Check if name is valid
    while True:
        name = input('Participant Name:')
        if name in aList:
            print(f'Sorry, {name} is already signed up in slot #{aList.index(name)+1}.')
        elif name.replace(' ','').isalpha() and any(x.isspace() for x in name):
            break
        else: print('\nInvalid input, please enter the participant\'s legel first name and last name and separate them by a white space.\n')
    #Check is slot is valid
    while True:
        slot = input(f'Desired starting slot #[1-{len(aList)}]:')
        if slot.isnumeric() == False:
            print('\nSorry, your input is invalid, please enter a postive integer\n')
        else:
            slotNum = int(slot)
            if slotNum > len(aList) or slotNum == 0:
                print('\nSorry, the slot is out of range. Please try again.\n')
            elif aList[slotNum-1] is not None:
                print(f'\nError:\nSlot #{slotNum} is filled. Please try again.\n')
            else: break
    aList[slotNum-1] = name
    print(f'\nSuccess:\n{name} is signed up in starting slot #{slotNum}.\n')
```

```python
def cancel_Sign_Up(aList):
    separator('Participant Cancellation')
    #Since it could take multiple attempts for a participant's sign up to be cancelled correcly, it has to have a while loop.
    while True:
        #check if the slot is valid
        while True:
            slot = input(f'Starting slot #[1-{len(aList)}]:')
            if slot.isnumeric() == False:
                print('\nSorry, your input is invalid, please enter a postive integer\n')
            else:
                slotNum = int(slot)
                if slotNum > len(aList) or slotNum == 0:
                    print('\nSorry, the slot is out of range. Please try again\n')
                else: break
        #check if the name is valid
        while True:
            name = input('Participant Name:') 
            if name.replace(' ','').isalpha() and any(x.isspace() for x in name):
                break
            else: print('\nInvalid input, please enter the participant\'s legel first name and last name and separate them by a white space.\n')
            break
            #check if the name is in the list
        if name not in aList:
            print(f'\nSorry, {name} is not in the participants\' list. Please try again.\n')
        elif name != aList[slotNum-1]:
            print(f'\nError:\n{name} is not in that starting slot.\n')
        else:
            aList[slotNum-1] = None
            print(f'\nSuccess:\n{name} has been cancelled from starting slot #{slotNum}.\n')
            break
                
```

```python
def view_Participants(aList):
    separator('View Participants')
    # check if the slot is valid
    while True:
        slot = input(f'Starting slot #[1-{len(aList)}]:')
        if slot.isnumeric() == False:
            print('\nSorry, your input is invalid, please enter a postive integer\n')
        else:
            slotNum = int(slot)
            if slotNum > len(aList) or slotNum == 0:
                print('\nSorry, the slot is out of range. Please try again\n')
            else: break
    print('\nStarting Slot: Participant')
    for i in range(max(1,slotNum-5),min(len(aList)+1,slotNum+6)):
        if aList[i-1] == None:
            print(f'{i}: [Empty]\n')
        else: print(f'{i}: {aList[i-1]}\n')
```

```python
def tournament_Tracker():
    separator('Welcome to Tournament R Us')
    participantList = []
    #Used to track if changes have been saved.
    changeSaved = True
    #Start Up UI
    while True:
        participantsNum = input('Enter the number of participants:')
        if participantsNum.isnumeric() == False:
            print('\nSorry, your input is invalid, please try to enter a positive integer.\n')
            continue
        else:
            numParticipants = int(participantsNum)
            participantList = [None] * numParticipants
            print(f'\nThere are {numParticipants} participant slots ready for sign-ups')
            break
    while True: 
        separator('\nParticipant Menu')
        option = input('1. Sign Up\n2. Cancel Sign Up\n3. View Participants\n4. Save Changes\n5. Exit\n')
        if option.isnumeric() == False:
            print('\nSorry, your input is invalid, please try to enter a positive integer in the list.\n')
            continue
        else:
            optionNum = int(option)
            if optionNum > 5 or optionNum == 0:
                print('\nSorry, the option you choose is not in the list. Please try again.\n')
                continue
        if optionNum == 1:
            sign_Up(participantList)
            changeSaved = False
        elif optionNum == 2:          
            cancel_Sign_Up(participantList)
            changeSaved = False
        elif optionNum == 3:
            view_Participants(participantList)
        elif optionNum == 4:
            separator('Save Changes')
            #Check if the user input is valid
            while True:
                #assign the variable globally so it can be used outside of the while loop
                global wantSave
                wantSave = input('Save your changes to CSV? [y/n]')
                if (wantSave != 'y' and wantSave != 'n'):
                    print('\nIn valid input. Please try again.\n')
                else: break
            #Check if user wants to save
            if wantSave == 'n':
                print('\nChanges are not saved if there is any. Return to the Main Menu.\n')
            elif changeSaved == True:
                print('\nAll changes have been saved or there is no change at all. There is no need to save. Return to the Main Menu.\n')
            else:
                with open(r'D:\Documents\Dev10\CourseNotes\Module06_Python_Basics\Tournament_Tracker\ParticipantList.csv', 'w') as savedFile:
                    writeFile = csv.writer(savedFile)
                    for i in range(len(participantList)):
                        if participantList[i] == None:
                            writeFile.writerow([f'{i+1}','[Empty]'])
                        else:
                            writeFile.writerow([f'{i+1}',f'{participantList[i]}'])
                changeSaved = True
        else:
            separator('Exit')
            if changeSaved == False:
                print('Any unsaved changes will be lost.')
            #Check if the user input is valid
            while True:
                #assign the variable globally so it can be used outside of the while loop
                global wantExit
                wantExit = input('Are you sure you want to exit? [y/n]')
                if (wantExit != 'y' and wantExit != 'n'):
                    print('\nIn valid input. Please try again.\n')
                else: break
            #Check if user wants to exit
            if wantSave == 'n':
                print('\nReturn to the Main Menu.\n')
            else:
                print('\nGoodbye!')
                break
```
